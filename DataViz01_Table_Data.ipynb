```python
df1= data[['sex','age','risk']]
df1
```

```python
set(df1['risk'])
```

```python
df1_1 = df1[df1['risk']=='อยู่ระหว่างสอบสวน']
df1_1
```

```python
df1_2 = df1[df1['risk']=='สนามมวย']
df1_2
```

```python
df1_3 = df1[df1['risk']=='สัมผัสผู้ติดเชื้อ']
df1_3
```

```python
df_merge = pd.concat([df1_1,df1_2,df1_3],ignore_index=True)
df_merge
```

```python
df_1 = df_merge['sex'].replace({
        'ชาย':20,
        'หญิง':60
    },inplace = True)
```
